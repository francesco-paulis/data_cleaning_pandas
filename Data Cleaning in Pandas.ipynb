### Cleaning/Standardizing phone numbers

```python
df
```

```python
# apllying the new format in the dataframe

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

```python
# deleting na and nan values

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','', regex = True) 
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','', regex = True) 

df
```

### Filtering down rows of data

```python
#I want a list of phone numbers that it's possible to contact, so is necessary 
#to exclude rows which in the 'do not contact' have the value 'yes' and the rows with no phone number.
```

```python
# filtering by column 'do not contact'; if there isn't any value, it's considered as a no

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
        
df

# Another way to drop null values:
# df = df.dropna(subset = "Phone_Number", inplace = True)
```
